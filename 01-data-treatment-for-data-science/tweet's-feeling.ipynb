## Exercise 1: Tweet's feeling
#### (written in Spanish)

Dado un archivo que contiene en cada línea una palabra o conjunto de palabras seguido de un valor numérico, denominado “sentimiento”, y un conjunto de tweets, se pide calcular para cada tweet un valor denominado “sentimiento del tweet”, que se obtiene como la suma de los “sentimientos” de los términos que aparecen en el tweet.

Obsérvese que:
- Aquellas palabras o conjunto de palabras que aparecen en un tweet, y no tengan un “sentimiento” asociado en el archivo, se contabilizarán como 0.
- Para procesar el archivo hay que tener en cuenta que cada palabra o conjunto de palabras están separadas de su valor mediante el carácter “\t”.
- Como resultado, se debe mostrar por pantalla un valor numérico en cada línea que represente el “sentimiento” de un tweet.
- No todos los tweets que se van a considerar tienen contenido, por lo que hay que filtrar aquellos que tienen de los que no tienen.
- El programa tendrá 2 parámetros de entrada: el archivo que contiene los sentimientos de los términos y el archivo que contiene los tweets.

```python
import numpy as np
import json  as js
```

```python
# Function which read the feelings file with name 'filename'.
# Return a dictionary with the words and its feeling.
def readFeelingsFile(filename) :
    data = open(filename)
    feelings = {}
    
    for line in data:
        word, feeling = line.split("\t")
        feelings[word] = int(feeling)
    
    return feelings
```

```python
# Function which read the tweets file with name 'filename'.
# Return a list with the tweets.
def readTweetsFile(filename) :
    data = open(filename)
    
    tweets = []
    
    for line in data :
        tweets.append(js.loads(line))
    
    return tweets
```

```python
# Function which gets the feeling of the 'text' param with the feelings in 'feelings' param.
def getTweetFeeling(feelings, text) :
    feeling = 0
    # For each word in tweet's text
    for word in text.split(" ") :
        if word in feelings : 
            # If the word is in feelings, its feeling's amount increments with its feeling in feelings
            feeling += feelings[word]
            
    return feeling
```

```python
# Function which calculates the tweet's feeling with the params: 'feelings' and 'tweets'.
# Return a list of string with the tweet's text and its feeling.
def calculateTweetsFeeling(feelings, tweets) :
    
    # Get the tweets with contents
    tweetsWithContents = [ tweet for tweet in tweets if 'delete' not in tweet ]
    
    tweetsFeeling = []
    
    # For each tweet
    for tweet in tweetsWithContents :
        # Get tweet's text
        text = tweet['text'] 
        # Get tweet's feeling
        feeling = getTweetFeeling(feelings, text)
        # Append the feeling
        tweetsFeeling.append(text + ' -- Feeling: ' + str(feeling))
            
    return tweetsFeeling
```

```python
if __name__ == '__main__' :
    
    # Get feelings of the file
    feelings = readFeelingsFile('feelings.txt')
    # Get tweets of the file
    tweets = readTweetsFile('tweets.txt')

    # Calculate tweet's feeling
    tweetsFeelings = calculateTweetsFeeling(feelings, tweets)
    
    # Show the results:
    for tweet in tweetsFeelings:
        print(tweet)
```
